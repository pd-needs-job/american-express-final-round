#### Q1. Find All Unique Pairs with Given Sum

#### Problem
Given an array of integers, find **all unique pairs** (not just one pair) that sum to a given target.  
Avoid duplicate pairs.

---

#### Example

**Input:**  nums = [1, 5, 3, 2, 4]  target = 6   **Output:** [(1, 5), (2, 4)]



```python
sample = [1, 5, 3, 2, 4]
target = 6


def find_pais(sample:list, target:int) ->list:
    visited_number = []      
    result_pair = []

    for number in sample:

        required_number = target - number  # Required number we need to find in corresponding to target number.

        if required_number in visited_number:

            pair = [number, required_number]    
            sorted_pair = sorted(pair)

            if sorted_pair not in result_pair:       # Avoiding Duplicates. (Like (1, 5) and (5, 1).
                result_pair.append(sorted_pair)
        
        visited_number.append(number)

    return result_pair


result = find_pais(sample, target)
print(result)

```

#### Q2. Most Frequent Fraction From Two Array

#### Problem
Given two arrays numerator[] and denominator[], compute fraction = numerator[i]/denominator[i] in reduced form for each i. 
Return the fraction that appears most frequently.

---

#### Example

**Input:**  nums = [1,2,3]  den=[2,4,6]   **Output:** 1/2

#### Pata nahi kya hi code hai ??? Samza ma hi nahi aaya !!

```python
from math import gcd  # Greatest Common Divisor (gcd)

def most_frequent_fraction(numerator_list:list, denominator_list: list):

    fraction_count = {}  # Stores Fraction String and How many times it appears.
    
    # zip() combines two (or more) lists element by element.It pairs the items at the same index together.
    for current_numerator, current_denominator in zip(numerator_list, denominator_list):  

        common_divisor = gcd(current_numerator, current_denominator) # Find the greatest common divisor of numerator and denominator.

        # reduce the fraction
        simple_numerator = current_numerator // common_divisor
        simple_denominator = current_denominator // common_divisor

        fraction_in_string_format = str(simple_numerator) + "/" + str(simple_denominator)  # Convert the fraction to string format for easy comparison.


        #. get() method is used to retrieve the value of a key from the dictionary. If the key does not exist, it returns the default value (0 in this case) instead of raising a KeyError.
        fraction_count[fraction_in_string_format] = fraction_count.get(fraction_in_string_format, 0) + 1 
        
    return max(fraction_count, key=lambda fraction: fraction_count[fraction])


print(most_frequent_fraction([1, 2, 3, 4], [2, 4, 6, 8]))  # Output: "1/2"
```

#### Q3. Calculate Precision, Recall and F1-Score from Scratch

#### Problem
Given actual labels and predicted labels (binary classification), compute precision, recall, and F1-score without using sklearn. This is crucial for fraud detection where class imbalance is extreme.

---

#### Example

**Input:**  Actual: [1,0,1,1,0,1]    Predicted: [1,0,0,1,0,1]     **Output:**    Precision=1.0,   Recall=0.75,   F1=0.857

```python
actual_output = [1,0,1,1,0,1]
predicted_output = [1,0,0,1,0,1]


def evaluate_model(actual_output: list, predicted_output:list) -> str:
    TP = FN = 0
    FP = TN = 0

    for actual, predicted in zip(actual_output, predicted_output):
        if actual == 1 and predicted == 1:  # Correctly Predicted Fraud
            TP += 1
        
        elif actual == 1 and predicted == 0:     # Missed Fraud
            FN += 1

        elif actual ==0 and predicted == 1:      # False Alarm
            FP += 1

        elif actual == 0 and predicted == 0:     # Correctly Ignored
            TN += 1



    if TP + FP  != 0:
        precision = TP / (TP + FP)
    else:
        precision = 0


    if TP + FN != 0:
        recall = TP / (TP + FN)
    else:
        recall =0

    

    if precision + recall != 0:
        F1_Score = 2 * precision * recall / (precision + recall)
    else:
        F1_Score = 0    


    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-Score:", F1_Score)


evaluate_model(actual_output, predicted_output)


```
