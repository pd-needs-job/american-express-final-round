##### Q1.  You are given a dataset containing predicted and actual values of a regression model. Compute the Mean Absolute Percentage Error (MAPE), which is widely used in model evaluation and risk validation.

```python
actual_list = [100, 200, 300, 400, 500]
predicted_list = [110, 190, 310, 390, 510]


def calculate_MAPE(actual_list:list , predicted_list: list) -> float:
    sum_of_all_percentages_error = 0
    for actual, predicted in zip(actual_list, predicted_list):
        if actual !=0 and predicted !=0:
            error = actual -predicted
            absolute_error = abs(error)
            absolute_percentage_error = (absolute_error / abs(actual)) * 100
            sum_of_all_percentages_error += absolute_percentage_error

    MAPE = sum_of_all_percentages_error / len(actual_list)

    return MAPE

result = calculate_MAPE(actual_list, predicted_list)
print(f"MAPE: {result:.2f}%")
```

##### Q3 Given a list of text prompts, remove duplicate prompts and return unique prompts. Useful in LLM prompt cleaning pipelines.

```python
prompts = ["hi", "hello", "hey", "hi","Yo", "hello", "yo"]


def remove_duplicates(prompts:list) -> list:

    unique_prompts = []

    for prompt in prompts:

        prompt_lower = prompt.lower()

        if prompt_lower not in unique_prompts:
            unique_prompts.append(prompt_lower)

    return unique_prompts

result = remove_duplicates(prompts)
print(result)
```

##### Q4. Given a list of sentences, count the frequency of each word. This is useful for NLP preprocessing.

```python
sentences_list = [
    "I love python",
    "I love coding",
    "python is fun"
]



def count_word_in_sentences(sentences_list: list) -> dict:

    word_count = {}

    for sentence in sentences_list:
        words_list = sentence.split()  # .split() gives list.

        for word in words_list:
            word_lower = word.lower()

            if word_lower in word_count:
                word_count[word_lower] += 1

            else:
                word_count[word_lower] = 1

    return word_count


result = count_word_in_sentences(sentences_list)
print(result)
```

##### Q5. Normalize a dataset column using Min-Max scaling (0 to 1). Important for ML model stability and fairness.

```python
# Normalization By Min Max = (X-Xmin)/(Xmax - Xmin)

arr = [10, 20, 30, 40, 50, -10, -20]

def min_max_normalization(arr: list) -> list:

    minimum_value = min(arr)
    maximum_value = max(arr)


    normalized_arr = []

    for value in arr:

        normalized_value = (value - minimum_value) / (maximum_value - minimum_value)
        normalized_arr.append(normalized_value)

    return normalized_arr

result = min_max_normalization(arr)
print(result)
```

##### Q6. Given predicted probabilities and a threshold, convert them into binary class predictions. This is often used in model evaluation pipelines.

```python
list_of_probabilites = [0.2, 0.7, 0.6, 0.9, 0.1]
threshold = 0.5

def classify_probabilities(list_of_probabilites: list, threshold: float) -> list:
    binary_class_predictions = []

    for probability in list_of_probabilites:

        if probability >= threshold:
            binary_class_predictions.append(1)
        else:
            binary_class_predictions.append(0)

    return binary_class_predictions

result = classify_probabilities(list_of_probabilites, threshold)
print(result)
```

##### Q7. Given a dataset, detect outliers using a simple threshold (e.g., values greater than mean + 2*std). Important for model risk and data quality checks.

```python
# threshold = mean + 2* standard deviation

import numpy as np

def identify_outliers(data: list) -> list:

    formatted_data = np.array(data)

    mean = np.mean(formatted_data)
    standard_deviation = np.std(formatted_data)

    threshold = mean + 2 * standard_deviation

    outliers = []

    for number in formatted_data:
        if number > threshold:
            outliers.append(int(number))

    return outliers

data = [10, 12, 14, 15, 18, 20, 100]
result = identify_outliers(data)
print(result)
```

##### Q8. Given a list of responses generated by an LLM, filter out responses that are too short (length < 5 words). Useful in GenAI evaluation pipelines.

```python
responses = [
    "hi",
    "this is a valid response",
    "hello",
    "how are you",
    "this works fine",
    "all good here",
    "response looks correct",
    "everything is okay",
    "nice to meet you",
    "valid input is received by the system"
]


def filter_small_responses(responses: list) -> list:

    filtered_responses = []

    for single_respons in responses:

        list_of_respons = single_respons.split()
        length_of_respons= len(list_of_respons)

        if length_of_respons >=5:
            filtered_responses.append(single_respons)
        
    return filtered_responses

result = filter_small_responses(responses)
print(result)
```

##### Q9. Find the intersection of two arrays.

```python
array1 = [1,2,2,1]
array2 = [2,2]


def intersection_of_arrays(array1: list, array2:list) -> list:

    intersections = []

    for element in array1:
        if element in array2 and element not in intersections:
            intersections.append(element)
    
    return intersections


result = intersection_of_arrays(array1, array2)
print(result)
```

##### Q10. Given a dataset, group values by a category and compute average.

```python
data = [("A", 10), ("A", 20), ("B", 30)]


def group_data(data:list) -> dict:
    grouped_data = {}

    for key, value in data:
        if key in grouped_data:
            grouped_data[key].append(value)

        else:
            grouped_data[key] =[value]

    return grouped_data


result = group_data(data)
print(result)

def calculate_mean_of_each(grouped_data: dict) -> dict:
    mean_values = {}

    for key, values in grouped_data.items():       # grouped_data.items() gives us key and list of values for that key
        mean_value = sum(values) / len(values)
        mean_values[key] = mean_value   

    return mean_values


grouped_data = group_data(data)
result = calculate_mean_of_each(grouped_data)
print(result)
```

##### Q11. Check if two strings are anagrams.

```python
def is_anagrams(string1:str, string2:str) -> bool:
    string1_lower = string1.lower()
    string2_lower = string2.lower() 
    sorted_string1 = sorted(string1_lower)
    sorted_string2 = sorted(string2_lower)
    result = sorted_string1 == sorted_string2
    return result

final = is_anagrams("listen", "silent")
print(final)
```

##### Q12. Given a list of transactions, find total spending per user.

```python
transaction_data = [("A", 100), ("B", 200), ("A", 50)]


def calculate_total_transaction_amount(transaction_data:list) -> dict:

    final_transaction= {}
    for key, value in transaction_data:

        if key in final_transaction:
            final_transaction[key] += int(value)

        else:
            final_transaction[key] = int(value)

    return final_transaction

result = calculate_total_transaction_amount(transaction_data)
print(result)
```
