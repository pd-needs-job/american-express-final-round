### Q1 Two Sum: Find two numbers that add to target.

```python
sample = [2, 7, 11, 15]
target = 9

def two_sum(sample: list, target: int) -> list:
    seen = {}

    for i in range(len(sample)):
        diff = target - sample[i]
        if diff in seen:
            return [seen[diff], i]
        seen[sample[i]] = i             # If particular number is not in seen then create a new dictionary with key as number from sample and value as index.
    seen[sample[i]] = i


two_sum(sample, target)
```

### Q4 Longest Substring Without Repeating Characters

### Longest Substring Without Repeating Characters

**Problem:**  
Given a string, find the **length** of the longest substring with **all unique characters**.

---

#### Key Points
- Substring = continuous sequence of characters  
- No character should repeat  
- Return **length**, not the substring

---

#### Example

Input: `"abcabcbb"`  
Output: `3`  
Explanation: `"abc"` is the longest substring without repeats

---

#### Intuition
- Traverse the string
- Keep a window of unique characters
- If duplicate appears → shrink window
- Track maximum length

---

#### Goal
Efficient solution using **Sliding Window + Set/HashMap** (O(n))

```python
def longest_substring(text: str) -> int:
    last_seen = {}
    start_index = 0
    longest_length = 0
    length_of_text = len(text)

    for current_index in range (length_of_text):

        current_char = text[current_index]

        if current_char in last_seen:
            start_index = max(start_index, last_seen[current_char] + 1)
    
        last_seen[current_char] = current_index
        longest_length = max(longest_length, current_index - start_index + 1)
    
    return longest_length

text = "abcabcbb"
result = longest_substring(text)
print(result)
```

### Q6 Move Zeroes

```python
sample_list = [3,4,0,17,6,0,9]

def move_zeroes(sample_list:list) -> list:
    result = []
    
    for number in  sample_list:
        if number !=0:
            result.append(number)

    length_of_result = len(result)
    length_of_sample_list = len(sample_list)
   
    while length_of_result < length_of_sample_list:
        result.append(0)
        length_of_result += 1

    return result

move_zero = move_zeroes(sample_list)
print(move_zero)
```

### Q7 Maximum Subarray (Kadane's Algorithm)

### Maximum Subarray (Kadane’s Algorithm)

**Problem:**  
Given an integer array `nums`, find the **maximum sum** of a **contiguous subarray**.

---

#### Key Points
- Subarray = continuous elements  
- Must contain at least one element  
- Return the **maximum possible sum**

---

#### Example

Input: `[-2,1,-3,4,-1,2,1,-5,4]`  
Output: `6`  
Explanation: Subarray `[4, -1, 2, 1]` has the largest sum

---

#### Goal
Efficient solution in **O(n)** time

```python
sample_list = [-2,1,-3,4,-1,2,1,-5,4]

def max_sub_array(sample_list: list) -> int:
    current_sum = 0
    max_sum = sample_list[0]

    for numb in sample_list:
        current_sum += numb

        if current_sum > max_sum:
            max_sum = current_sum

        elif current_sum < 0:
            current_sum = 0
    return max_sum


result = max_sub_array(sample_list)
print(result)
```

### Q8 Detect Cycle in Linked List

👉 Asked in AmEx interview

#### 🧠 Detect Cycle in Linked List (AmEx Interview)

#### 📌 Problem Statement
Given the head of a linked list, determine whether it contains a **cycle (loop)**.

A cycle exists if a node’s `next` pointer eventually points back to a previous node.

Example:


---

#### 🚀 Approach: Floyd’s Cycle Detection Algorithm (Tortoise & Hare)

Use two pointers:
- `slow` → moves 1 step at a time
- `fast` → moves 2 steps at a time

---

#### 💡 Key Idea
- If a cycle exists → fast and slow pointers will eventually meet
- If no cycle → fast reaches `None`

---


```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
    

def detect_cycle(head: Node) -> bool:
    first_pointer = head            # moves slowly (1 step)
    second_pointer = head           # moves fast (2 steps)

    while second_pointer and second_pointer.next:
        
        # move pointers
        first_pointer = first_pointer.next
        second_pointer = second_pointer.next.next

        # if they meet, cycle exists
        if first_pointer == second_pointer:
            return True
        
    # if loop ends, no cycle
    return False
```

```python
# Create nodes
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)

# Connect nodes (1 → 2 → 3 → 4)
a.next = b
b.next = a
#c.next = d

# Test
print(detect_cycle(a))   # Expected: True
```

#### Q9 Reverse Linked List

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None


def reverse_list( head: Node) -> Node:
    previous_node = None
    current_node = head

    # Arrow(.next) always moves from left to right.

    while current_node:
      # First we are changing the direction of the arrows of each node.
      saving_next_node_value = current_node.next   #Technically we are storing next node value which woulf be none # Saves next node. Here it is none. 

      # Reverse the arrow direction. (From current to previous)
      current_node.next = previous_node
      
      # Now we are interchanging the values of current and previous node.
      previous_node = current_node   
      current_node = saving_next_node_value

    return previous_node
```

```python
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4) 


node1.next = node2
node2.next = node3      
node3.next = node4  

print("Reverse Node")
reversed_head = reverse_list(node1)
print(reversed_head.value)  # Should print 4
```

#### Q10 Implementing Stack Using List

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)

    def pop(self):
        if self.stack:
            return self.stack.pop()
        
    def top(self):
        top = self.stack[-1]
        return top
```

```python
# ✅ Example to check
s = Stack()

s.push(10)
s.push(20)
s.push(30)

print("Top element:", s.top())   # 30

print("Popped:", s.pop())       # 30
print("Top after pop:", s.top()) # 20

print("Popped:", s.pop())       # 20
print("Popped:", s.pop())       # 10

print("Popped from empty:", s.pop())  # None
```

#### Q11 12. Merge Sort (VERY IMPORTANT)

```python
sample = [5,2,3,1]

def merge_sort(sample:list) -> list:

    # If list has single element or is empty.
    if len(sample) <= 1:
        return sample
    
    mid = len(sample) // 2
    left_half = merge_sort(sample[:mid])     # Recusrsion
    right_half = merge_sort(sample[mid:])    # Recursion
    
    result = []
    i = j = 0

    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            result.append(left_half[i])
            i += 1

        else:
            result.append(right_half[j])
            j += 1

    return (result + left_half[i:] + right_half[j:])



sorted_list = merge_sort(sample)
print(sorted_list)

    
        
```
